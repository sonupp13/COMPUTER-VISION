```python
import numpy as np
import cv2
from google.colab.patches import cv2_imshow

# We point OpenCV's CascadeClassifier function to where our
# classifier (XML file format) is stored
face_classifier = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

# Load our image then convert it to grayscale
image = cv2.imread(r'/content/drive/MyDrive/Colab Notebooks/623867ee70502_download[628].jpg')
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)


# Our classifier returns the ROI of the detected face as a tuple
# It stores the top left coordinate and the bottom right coordiantes
faces = face_classifier.detectMultiScale(image, scaleFactor=1.3, minNeighbors=5)


#faces will contain a list of rectangles representing the detected faces in the image.
#Each rectangle contains the coordinates (x, y, w, h) of the detected face,
# where (x, y) is the top-left corner of the rectangle, and (w, h) is its width and height.


# When no faces detected, face_classifier returns and empty tuple
if faces is():
    print("No faces found")

# We iterate through our faces array and draw a rectangle
# over each face in faces
for (x,y,w,h) in faces:
    cv2.rectangle(image, (x,y), (x+w,y+h), (255,255,255), 2)
    cv2_imshow(image)
    cv2.waitKey(0)

cv2.destroyAllWindows()
```
